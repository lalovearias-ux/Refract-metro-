```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import time
import random

# --- 1. DEFINIR LOS 6 EQUIPOS ---
equipos = {
    1: {"nombre": "Leones 🦁", "color": "#FF4136", "texto": "blanco"},
    2: {"nombre": "Tiburones 🦈", "color": "#0074D9", "texto": "blanco"},
    3: {"nombre": "Dinosaurios 🦖", "color": "#2ECC40", "texto": "blanco"},
    4: {"nombre": "Rayos ⚡", "color": "#FFDC00", "texto": "negro"},
    5: {"nombre": "Monstruos 👾", "color": "#B10DC9", "texto": "blanco"},
    6: {"nombre": "Zorros 🦊", "color": "#FF851B", "texto": "blanco"}
}

# --- 2. CREAR LA INTERFAZ ---
titulo = widgets.HTML("<h1 style='text-align: center; font-size: 40px; color: #333;'>🎲 El Dado Mágico 🎲</h1>")
instruccion = widgets.HTML("<h3 style='text-align: center;'>¡Toca el botón para saber tu equipo!</h3>")

boton_dado = widgets.Button(
    description='🎲 ¡TIRAR EL DADO! 🎲',
    button_style='info',
    layout=widgets.Layout(width='400px', height='80px')
)
boton_dado.style.font_weight = 'bold'
boton_dado.style.button_color = '#17a2b8'

area_resultado = widgets.Output()

# --- 3. EL MOTOR DE ANIMACIÓN RÁPIDA ---
def tirar_dado(b):
    boton_dado.disabled = True

    with area_resultado:
        # Animación rápida de ruleta (10 cuadros rapidísimos)
        for _ in range(10):
            clear_output(wait=True)
            equipo_al_azar = random.choice(list(equipos.values()))
            html_ruleta = f"""
            <div style='background-color: #f4f4f4; padding: 20px; border-radius: 20px; text-align: center; border: 5px solid #ccc;'>
                <h1 style='font-size: 80px; margin: 0;'>{equipo_al_azar["nombre"]}</h1>
            </div>
            """
            display(HTML(html_ruleta))
            time.sleep(0.08) # Pausa cortísima

        # El resultado final
        clear_output(wait=True)
        numero_ganador = random.randint(1, 6)
        equipo_final = equipos[numero_ganador]

        resultado_html = f"""
        <div style='background-color: {equipo_final["color"]}; padding: 30px; border-radius: 20px; text-align: center; color: {equipo_final["texto"]}; border: 8px solid #333; box-shadow: 5px 5px 15px rgba(0,0,0,0.3);'>
            <h2 style='font-size: 35px; margin:0; text-transform: uppercase;'>Equipo Número {numero_ganador}</h2>
            <h1 style='font-size: 90px; margin: 10px 0;'>{equipo_final["nombre"]}</h1>
        </div>
        """
        display(HTML(resultado_html))

    boton_dado.disabled = False # Reactivamos el botón para el siguiente niño

boton_dado.on_click(tirar_dado)

# --- 4. MOSTRAR TODO EN PANTALLA ---
contenedor_centrado = widgets.VBox(
    [titulo, instruccion, boton_dado, widgets.HTML("<br>"), area_resultado],
    layout=widgets.Layout(align_items='center')
)

display(contenedor_centrado)
```
